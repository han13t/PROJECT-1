# Data analysis of the dataset of all inventories of red wine #

```python
import pandas as pd
import seaborn as sn
from matplotlib import pyplot as plt
import numpy as np
from scipy import stats
```

```python
data = pd.read_csv('data.csv')
data
```

# __Now we need to describe the dataset to identify and pre-process any null and duplicate values.__ #

```python
data.info()
```

##  An ordered way to identify null values is to:  ##

```python
data_summary = data.describe()

null_counts = data.isnull().sum()

data_summary.loc['null_count'] = null_counts

data_summary = data_summary.transpose()

data_summary
```

__The isnull().sum() method is used to count the number of null values in each columns. We then add the null count information to the data_summary statistics using the loc[ ] method and we create a new row called the "null.count".Then finally we use the transpose() method to allow easy viewing so that the columns are the variables and the rows are the data_summary statistics.
From the data summary we see the the columns that contain null values.__

## In order to pre- process the dataset we need to rename some of the columns in the dataset.  To do so we do the following: ##

```python
data = data.rename(columns={'fixed acidity':'fixed_acidity', 'volatile acidity':'volatile_acidity', 'citric acid':'citric_acid', 'residual sugar':'residual_sugar', 'free sulfur dioxide':'free_sulfur_dioxide', 'total sulfur dioxide':'total_sulfur_dioxide'})
data
```

## _We can choose to get rid of the null values for good and to do so we do the following:_ ##

```python
x = data.dropna()
x
```

__The .dropna() introduced removes instances where null sets are found and leaves instances that have value. Therefore as seen above the number of rows decreases from 1601 to 1591 but we don't want to reduce the number of rows so we won't use the dropna method__.

## Pre-Processing using the mean method ##

__A better way of replacing null values is by finding its mean,mode or median. We choose to use the mean method because it gives us the average value of the columns that have null values.__

```python
data
```

```python
mean_fixed_acidity = data.fixed_acidity.mean()
mean_citric_acid = data.citric_acid.mean()
mean_chlorides = data.chlorides.mean()
mean_free_sulfur_dioxide = data.free_sulfur_dioxide.mean()
mean_total_sulfur_dioxide = data.total_sulfur_dioxide.mean()
mean_density = data.density.mean()
mean_pH = data.pH.mean()
mean_quality = data.quality.mean()
```

```python
print(f'fixed_acidity: {mean_fixed_acidity}')
print(f'citric_acid: {mean_citric_acid}')
print(f'chlorides: {mean_chlorides}')
print(f'free_sulfur_dioxide: {mean_free_sulfur_dioxide}')
print(f'total_sulfur_dioxide: {mean_total_sulfur_dioxide}')
print(f'density: {mean_density}')
print(f'pH: {mean_pH}')
print(f'quality: {mean_quality}')
```

```python
data['fixed_acidity'] = data['fixed_acidity'].fillna(mean_fixed_acidity)
data['citric_acid'] = data['citric_acid'].fillna(mean_citric_acid)
data['chlorides'] = data['chlorides'].fillna(mean_chlorides)
data['free_sulfur_dioxide'] = data['free_sulfur_dioxide'].fillna(mean_free_sulfur_dioxide)
data['total_sulfur_dioxide'] = data['total_sulfur_dioxide'].fillna(mean_total_sulfur_dioxide)
data['density'] = data['density'].fillna(mean_density)
data['pH'] = data['pH'].fillna(mean_pH)
data['quality'] = data['quality'].fillna(mean_quality)
```
